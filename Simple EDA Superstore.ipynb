```python
data['order_year']=data.order_date.dt.strftime('%Y')
data[:]
```

<p style='color:blue'>Dapat dilihat bahwa dalam data tsb customer_id masih banyak yang dobel, jadi perlu untuk dijadikan 1 baris hanya 1 customer_id yang berbeda (tiap tahunnya)

```python
data5= data.groupby(['order_year','customer_id']).agg({
    'sales':'sum',
    'profit':'sum'
})

data5 = data5.sort_values('order_year')
data5 =data5.reset_index()
data5
#kalau len(data5.customer_id.unique()) saja ada = 793 
#tapi hasil groupby ada 2499 baris soalnya tiap tahun juga keitung, jadi customer_id bisa dobel
```

```python
data5.order_year.unique()
```

```python
tahun=np.array(data5['order_year'])
a,b,c,d=0,0,0,0
for i in range (0,len(tahun)):
    if tahun[i]=='2014':
        a=a+1
    elif tahun[i]=='2015':
        b=b+1
    elif tahun[i]=='2016':
        c=c+1
    elif tahun[i]=='2017':
        d=d+1
print(a,b,c,d)
print(f'2014, index 0 sampai ke {a-1}')  
print(f'2015, index {a} sampai ke {b+a-1}')
print(f'2016, index {b+a} sampai ke {c+b+a-1}')
print(f'2017, index {c+b+a} sampai ke {d+c+b+a-1}')
```

# Membuat visualisasi

```python
import plotly
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
```

```python
data = data.sort_values('order_year')
data = data.reset_index()
data
```

<p style='color:blue'>Misal saya akan membuat 'Sales vs Profit' hanya untuk tahun 2014 dan 2015 saja

```python
data[4092:4096]
#data tahun 2014-2015 index [0:4094]  
```

```python
data7 = data[:4094]
data7
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(x=data7.sales, y=data7.profit)
lines = slope * data7.sales.values + intercept
lines
```

```python
dot1 = go.Scatter(
    x=data7[data7.category=='Furniture'].sales,
    y=data7[data7.category=='Furniture'].profit,
    mode='markers',
    marker={
        'size':10
    },
    name='Furniture'
)

dot2 = go.Scatter(
    x=data7[data7.category=='Office Supplies'].sales,
    y=data7[data7.category=='Office Supplies'].profit,
    mode='markers',
    marker={
        'size':5
    },
    name='Office Supplies'
)

dot3 = go.Scatter(
    x=data7[data7.category=='Technology'].sales,
    y=data7[data7.category=='Technology'].profit,
    mode='markers',
    name='Technology'
)

line= go.Scatter(
    x=data7.sales,
    y=lines,
    line={
        'dash':'dash',
        'width': 5
    },
    name='Linear Trend Line'
)


data=[dot1, dot2, dot3, line]

layout = {
    'title': {
        'text' : 'Sales VS Profit 2014-2015',
        'x' : 0.5 # nilai x antara 0 - 1, jadi 0.5 agar title ditengah
    },
    
    'xaxis': { # memberi label
        'title':'Sales',
        'range' : [0,7500] # membatasi nilai 
    },
    'yaxis': {
        'title':'Profit',
        'range' : [-2500,2500]
        
    },
    
    'legend': {  # mengubah posisi legend 
        'orientation' : 'h', #jadi horizontal dibawah
        'x':0.5,
        # 0,5 kiri ;  1 kanan  (atas)
        'y':1,#jadi horizontal diatas
        'xanchor':'center',#horizontal diatas tengah
    }
    
}

fig=go.Figure(data=data,layout=layout)
iplot(fig)
```
